```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'b--')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'b')
axes[1].plot(y, x, 'b--')
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4)
axes[0, 0].plot(x, np.cos(x))
axes[0, 1].plot(x, np.sin(x), 'y')
axes[0, 2].plot(x, np.tan(x), 'r')
axes[0, 3].plot(x, np.cos(x)**2, 'b--')
```

```python
fig, ((ax1, ax2, ax3, ax4), (ax5, ax6, ax7,
                            ax8) )= plt.subplots(nrows=2, ncols=4)
ax1.plot(x, np.cos(x))
ax2.plot(x, np.sin(x), 'y')
ax3.plot(x, np.tan(x), 'r')
ax4.plot(x, np.cos(x)**2, 'b--')
fig.tight_layout()
```
