```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
```

```python
print(plt.style.available)
```

```python
plt.style.use('classic')
```

```python
fig, axes = plt.subplots(figsize=(8, 8))

axes.plot(x, x+1, 'b--')
axes.plot(x, x+2, 'r-')
axes.plot(x, x+3, 'yo:')
```

# pyplot

```python
fig, axes = plt.subplots(figsize=(8, 8))

axes.plot(x, x+1, color='green', alpha=0.8, linewidth=5, linestyle='-', marker='o')
axes.plot(x, x+2, color='green')
axes.plot(x, x+3, color='blue')
```
