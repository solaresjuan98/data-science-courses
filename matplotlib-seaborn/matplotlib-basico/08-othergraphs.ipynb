```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1, 50, 100)
```

# Histograma

```python
plt.hist(data, bins=10, width=10, histtype='bar')
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

# Boxplot

```python
plt.boxplot(data)
plt.show()
```

```python
# horizonal
plt.boxplot(data, vert=False, patch_artist=True)
plt.show()
```

```python
# 
data = np.append(data, 200)
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```
