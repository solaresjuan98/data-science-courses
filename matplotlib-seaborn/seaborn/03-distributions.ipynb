```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# * Acumulated
sns.histplot(data = tips, x = 'tip', bins=15, cumulative=False, hue='sex', stat='count', multiple='dodge')
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', shade=True)
plt.show()
```

```python
sns.kdeplot(data=tips, x='tip', hue='sex', cumulative=True, shade= True)
plt.show()
```

```python
sns.ecdfplot(data=tips, x='tip', hue='sex', stat='count')
plt.show()
```

```python
sns.displot(data=tips, x='tip', hue='sex',  kind='hist', multiple='stack')
plt.show()
```
